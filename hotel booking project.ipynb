# DATA ANALYSIS AND VISUALIZATION


```python
canceled_perc = df['is_canceled'].value_counts(normalize = True)
print(canceled_perc)

plt.figure(figsize = (5,4))
plt.title('Reservation status count')
plt.bar(['Not Canceled','Canceled'],df['is_canceled'].value_counts(),edgecolor = 'k')
plt.show()
```

```python
ax = sns.countplot(data= df, x="hotel", hue = "is_canceled")

plt.title("Reservation of different hotels")
plt.xlabel('hotel')
plt.ylabel('number of reservation')
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title("Average Daily Rate in City and Resort Hotel", fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label= "Resort Hotel")
plt.plot(city_hotel.index, city_hotel['adr'], label= "City Hotel")
plt.legend()
plt.show()
```

Resort hotel has high ADR as compaire to City hotel.

```python
df['Month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (20,8))
ax1 = sns.countplot(x='Month' , hue = 'is_canceled' , data=df)
plt.legend(['not canceled' , 'canceled'])
plt.ylabel("No of reservations")
plt.show()
```

From above analysis in "January" hotels got max cancelation and min in "August".

```python
plt.figure(figsize=(15,8))
plt.title("ADR ", fontsize = 20)

sns.barplot(x='Month', y='adr', data=df[df['is_canceled'] == 1].groupby('Month')[['adr']].sum().reset_index())

plt.show()
          
          
```

In January , adr is max and in August , adr is min.
When the adr is high , cancelation also high (in January) and whem adr is low , cancelation is also low(in August).

```python
cancelation_data = df[df['is_canceled']==1]
top_ten_country = cancelation_data['country'].value_counts().head(10)
plt.title("top 10 countries with reservation canceled")
plt.pie(top_ten_country,autopct = "%0.2f", labels = top_ten_country.index)
plt.show()
```

Country "PRT" has max cancelation, so we need to pay more attention in this country's hotel facilities.
